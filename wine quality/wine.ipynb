### Feature Selection

```python
df.sample(5)
```

```python
df['quality'].unique()
```

```python
df['goodquality']=[1 if x>=7 else 0 for x in df['quality']]
df.sample(5)
```

```python
df.goodquality.value_counts()
```

```python
X=df.drop(['quality','goodquality'],axis=1)
Y=df['goodquality']
```

```python
X.shape
Y.shape
```

#### Feature Importance

```python
from sklearn.ensemble import ExtraTreesClassifier
classification_tree= ExtraTreesClassifier()
classification_tree.fit(X,Y)
score= classification_tree.feature_importances_
print(score)
```

#### Spliting dataset

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=16)
```

```python
model_res= pd.DataFrame(columns=['Models','Score'])
```

```python
from sklearn.linear_model import  LogisticRegression
model= LogisticRegression()
model.fit(X_train, y_train)
y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import  accuracy_score,precision_score,r2_score,confusion_matrix
model_res.loc[len(model_res)]=['recall',r2_score(y_test,y_pred)]
model_res
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
model_res.loc[len(model_res)]=['Knn',accuracy_score(y_test,y_pred)]
model_res
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
model_res.loc[len(model_res)]=['SVC',accuracy_score(y_test,y_pred)]
model_res
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
model_res.loc[len(model_res)]=['Decision',accuracy_score(y_test,y_pred)]
model_res
```

```python
from sklearn.ensemble import RandomForestClassifier
r_boost = RandomForestClassifier()
r_boost.fit(X_train, y_train)
y_pred = r_boost.predict(X_test)
model_res.loc[len(model_res)]=['Randomforest',accuracy_score(y_test,y_pred)]
model_res
print(confusion_matrix(y_test,y_pred))
print(f'Recall: {r2_score(y_test,y_pred)}')
print(f"Precision:{precision_score(y_test,y_pred)}")
# print(f"f1_score: {f1_score(y_test,y_pred)}")
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
model_res.loc[len(model_res)]=['Gnb',accuracy_score(y_test,y_pred)]
model_res
```

```python
import xgboost as xgb
model = xgb.XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)

model_res.loc[len(model_res)]=['Xgb',accuracy_score(y_test,y_pred)]
model_res
```
